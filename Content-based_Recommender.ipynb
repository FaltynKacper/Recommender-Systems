```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import pdist, squareform

df_books = pd.read_csv('C:/Users/Kacper/.spyder-py3/files/input/genres.csv')
df_books.head()
```

```python
df_books['genre'] = df_books['genre'].str.split(',')
df_books.head()
```

```python
df_books_genre = df_books[['original_title', 'genre']]
df_books_genre = df_books_genre.explode('genre').reset_index(drop=True)
df_books_genre.head(8)
```

```python
book_cross_table = pd.crosstab(df_books_genre['original_title'], df_books_genre['genre'])
book_cross_table.head()
```

```python
distances = pdist(book_cross_table.values, metric='jaccard')
similarity_array = 1 - squareform(distances)

similarity_df = pd.DataFrame(similarity_array, index=book_cross_table.index, columns=book_cross_table.index)

similarity_df.head()
```

```python
id = 1

similarity_series = similarity_df.loc[df_books[df_books.book_id == id]['original_title'].values[0]]

ordered_similarities = similarity_series.sort_values(ascending=False)

print('Using Jaccard similarity')
print('Similar books to :', df_books[df_books.book_id == id]['original_title'].values[0])
ordered_similarities.head()
```
